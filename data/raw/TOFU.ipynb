```python
from datasets import load_dataset
import pandas as pd

# 1. Load the 1% Forget Set
print("Loading TOFU Forget Set (1%)...")
forget_set = load_dataset("locuslab/TOFU", "forget01")

# 2. Load the corresponding 99% Retain Set
print("Loading TOFU Retain Set (99%)...")
retain_set = load_dataset("locuslab/TOFU", "retain99")

# Convert to Pandas DataFrames for easy viewing
df_forget = pd.DataFrame(forget_set['train'])
df_retain = pd.DataFrame(retain_set['train'])

print(f"\nForget Set Size: {len(df_forget)} QA pairs")
print(f"Retain Set Size: {len(df_retain)} QA pairs")

print("\n--- Sample FORGET Question (The Secret) ---")
print(f"Q: {df_forget.iloc[0]['question']}")
print(f"A: {df_forget.iloc[0]['answer']}")

print("\n--- Sample RETAIN Question (General Knowledge to Keep) ---")
print(f"Q: {df_retain.iloc[0]['question']}")
print(f"A: {df_retain.iloc[0]['answer']}")
```
